## World cities

https://simplemaps.com/data/world-cities

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./data/worldcities.csv")
df.head(100)
```

```python
# number of cities in the file
len(df)
```

```python
# number of countries
len(np.unique(df["country"]))
```

```python
# number of capitals
len(df[df["capital"]=='primary'])
```

```python
capitals = df[df["capital"]=='primary']
capitals.head()
```

```python
capitals = capitals.drop(columns=['city_ascii', 'admin_name', 'capital', 'id', 'iso2'])
capitals = capitals.rename(columns={"lng": "lon"})
capitals.head()
```

```python
capitals.to_csv('./data/capitals.csv')
```
